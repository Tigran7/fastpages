# Basics

## Print Line
System.out.println()
Prints to the console
- System:class from the core java library
- out: an object that controls the output
- println(): a method associated with the object that receives a single argumente, shortened version of print line

```java
System.out.println("Hello World!");
```

## Comments
Text that is ignored by the compiler, used to increase program readability
- Single line comment //
- Multi line comment /* */


```java
// Single line comment
 
/*
This is a 
multi-line comment!
*/
```

## Main() method
- Every java application must contain a main() method
- main() invokes all other methods
- Typically seen as public static void main(String[] args) { }
    - It accepts a single argument

```java
public class Factory {
  
  public static void main(String[] args) {
    // the statement above uses the main() method
    System.out.println("Welcome to the Chocolate Factory!");
 
  }
  
}
```

## Classes
- represents a single concept
- A Java program must have one class whose name is the same as the program filename

```java
// Person class must be declared in a program file named Person.java
public class Person {
  
  public static void main(String[] args) {
    
    System.out.println("I am a person, not a computer.");
    
  }
  
}
```

## Compiling Java
Compiling a program in java works like this:
1. Each individual class converted into .class file (byte code)
2. JVM (Java Virtual machine) runs byte code
- Compiling in Java helps us catch errors, a file with errors won't be compiled

To compile a file named MyProgram.java, in terminal run:
1. javac MyProgram.java
    - This creates a .class file, MyProgram.class
2. Execute class file: java MyProgram
    - If there are errors in the code a class file will not be created

## Whitespace
Whitespace, including spaces and newlines, between statements is ignored.

```java
System.out.println("Example of a statement");
 

System.out.println("Another statement");
 
// Output:
// Example of a statement
// Another statement
```

## Statements
Line of code that executes a task and is terminated with a ; (semicolon)

```java
System.out.println("This is a statement");
```
